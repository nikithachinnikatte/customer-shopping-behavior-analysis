```python
import pandas as pd

df = pd.read_csv("data/customer_shopping_behavior.csv")

df.head()
```

```python
df.info()#gives us column name
```

```python
df.describe()#summary statistics of only numerical columns
```

```python
df.describe(include="all")# summary statistics of all columns
```

```python
df.isnull().sum()#checks the missing values and choose median as a solution for null since median r robust to outliers whereas mean are not
```

```python
#find the median value within each category n replace that with the final missing value since herethe review rating has 37 nulls
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()#now there will be no null values
```

```python
#creating snake case for all columns names for uniformity
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amt'})
```

```python
df.columns
```

```python
#create a common age grp
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase frequency days
#create a dictionary first
# clean column
df['frequency_of_purchases'] = df['frequency_of_purchases'].str.strip().str.lower()
df = df.dropna(subset=['purchase_frequency_days','frequency_of_purchases'], how='all')

# mapping dictionary
frequency_mapping = {
    'fortnight':14,
    'week':7,
    'weekly':7,
    'monthly':30,
    'quarterly':90,
    'biweekly':14,
    'annually':365,
    'every 3 months':90
}

# create new column
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

# check results
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df = df.reset_index(drop=True)#considers all rows without any conditions applied 
df[['discount_applied','promo_code_used']].head(10)
```

```python
bool((df['discount_applied']==df['promo_code_used']).all())
```

```python
df=df.drop('promo_code_used',axis=1)
```
